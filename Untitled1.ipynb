```python
import asyncio
import nest_asyncio
import sys
from playwright.async_api import async_playwright

# 1. Apply the patch to allow nested event loops in Jupyter
nest_asyncio.apply()

# 2. Force the Proactor policy for Windows subprocess support
if sys.platform == 'win32':
    asyncio.set_event_loop_policy(asyncio.WindowsProactorEventLoopPolicy())

async def scrape_capcut(url):
    try:
        async with async_playwright() as p:
            # We use chromium; headless=True keeps it hidden
            browser = await p.chromium.launch(headless=True)
            
            # Set a common user agent to look less like a bot
            context = await browser.new_context(
                user_agent="Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36"
            )
            page = await context.new_page()
            
            print(f"Opening {url}...")
            await page.goto(url, wait_until="domcontentloaded", timeout=60000)
            
            # CapCut loads content dynamically; wait 3 seconds for JS to run
            await asyncio.sleep(3) 
            
            content = await page.content()
            print(f"Success! Captured {len(content)} characters of HTML.")
            
            await browser.close()
    except Exception as e:
        print(f"An error occurred: {e}")

# 3. Execution
await scrape_capcut("https://www.capcut.com/")
```
